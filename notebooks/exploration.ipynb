# Brain Tumor Segmentation – Exploration Notebook

This notebook is used for **data exploration, visualization, and sanity checks** for the MRI brain tumor segmentation project.

**Dataset:** BraTS (MRI – NMR based imaging)

**Models:** U-Net, Attention U-Net


```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

## Load and Visualize an MRI Volume
This section loads a `.nii/.nii.gz` MRI file and visualizes a middle slice.

```python
# Update path to your MRI file
mri_path = '../data/sample_mri/sample.nii.gz'

img = nib.load(mri_path)
data = img.get_fdata()

slice_idx = data.shape[2] // 2
plt.imshow(data[:, :, slice_idx], cmap='gray')
plt.title('Middle Slice of MRI')
plt.axis('off')
plt.show()
```

## Intensity Normalization
MRI intensities vary across scans. Z-score normalization is applied.

```python
slice_img = data[:, :, slice_idx]
norm_img = (slice_img - np.mean(slice_img)) / (np.std(slice_img) + 1e-8)

plt.imshow(norm_img, cmap='gray')
plt.title('Normalized MRI Slice')
plt.axis('off')
plt.show()
```
